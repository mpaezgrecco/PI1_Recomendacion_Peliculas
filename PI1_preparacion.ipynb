Algunos campos, como belongs_to_collection, production_companies y otros (ver diccionario de datos) estÃ¡n anidados, esto es o bien tienen un diccionario o una lista como valores en cada fila, Â¡deberÃ¡n desanidarlos para poder y unirlos al dataset de nuevo hacer alguna de las consultas de la API! O bien buscar la manera de acceder a esos datos sin desanidarlos.

```python
# funciÃ³n para convertir un diccionario a string y extraer el valor de "name"
def parse_and_extract_name(record):
    if record is None:
        return None
    if isinstance(record, dict):
        return record.get('name')
    # Si record es una cadena que parece un diccionario, evalÃºa y extrae "name"
    if isinstance(record, str):
        try:
            record_dict = eval(record)
            if isinstance(record_dict, dict):
                return record_dict.get('name')
        except:
            return None
    return None

# Aplicar la funciÃ³n a la columna "belongs_to_collection" y crear la nueva columna "name_to_collection"
data['name_to_collection'] = data['belongs_to_collection'].apply(parse_and_extract_name) 
```

```python
def desanidar_name(columna, lugar_data):
    # lugar_data indica la posicion del disccionario donde esta ubicado name
    global data

    # Convertir strings JSON en listas de diccionarios
    data[columna] = data[columna].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else [])

    # Extraer los nombres en la posiciÃ³n especificada y almacenarlos en una nueva columna
    data[f'name_{columna}'] = data[columna].apply(
        lambda x: [dic.get('name') for dic in x] if isinstance(x, list) else []
    ).apply(lambda x: x[lugar_data] if len(x) > lugar_data else None)
```

Los valores nulos de los campos revenue, budget deben ser rellenados por el nÃºmero 0.

```python
# Rellenar valores nulos de las columnas 'revenue' y 'budget' con 0
data['revenue'] = data['revenue'].fillna(0)
data['budget'] = data['budget'].fillna(0)
```

Los valores nulos del campo release date deben eliminarse.

```python
data = data.dropna(subset=['release_date'])
```

De haber fechas, deberÃ¡n tener el formato AAAA-mm-dd, ademÃ¡s deberÃ¡n crear la columna release_year donde extraerÃ¡n el aÃ±o de la fecha de estreno.

```python
# Convertir la columna 'release_date' al formato de fecha AAAA-mm-dd
data['release_date'] = pd.to_datetime(data['release_date'], errors='coerce', format='%Y-%m-%d')
```

```python
# Eliminar filas con fechas no vÃ¡lidas (que no pudieron convertirse)
data = data.dropna(subset=['release_date'])
```

```python
# Extraer el aÃ±o de la fecha de estreno y crear la nueva columna 'release_year'
data['release_year'] = data['release_date'].dt.year
```

Crear la columna con el retorno de inversiÃ³n, llamada return con los campos revenue y budget, dividiendo estas dos Ãºltimas revenue / budget, cuando no hay datos disponibles para calcularlo, deberÃ¡ tomar el valor 0.

```python
# Convertir 'budget' a tipo float
data['budget'] = data['budget'].astype(float)
```

```python
# Crear la columna 'return' con el cÃ¡lculo de revenue / budget
data['return'] = np.where((data['budget'] == 0) | (data['revenue'] == 0) | (data['budget'].isna()) | (data['revenue'].isna()), 
                          0, 
                          data['revenue'] / data['budget'])
```
